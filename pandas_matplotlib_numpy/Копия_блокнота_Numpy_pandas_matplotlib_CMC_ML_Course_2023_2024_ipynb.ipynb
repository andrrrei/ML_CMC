```python
data = data.explode('tags')
data.loc[data['tags'] == '[none]', 'tags'] = ''
matrix = data['tags'].apply(lambda x: x.strip().lower().split('|'))
merged_function = lambda matrix: [el for lst in matrix for el in lst if el != '']
merged = (merged_function(matrix))
from collections import Counter
d = Counter(merged)
next(iter({k: v for k, v in sorted(d.items(), key=lambda item: item[1], reverse = True)}))
```

```python
import time
import numpy as np

import functions_vectorised
import functions
```

```python
# генерация данных

sum_non_neg_diag_data = []
for n in [10, 100, 1000, 10000]:
    X = np.random.randint(0, 3, (n, n))
    sum_non_neg_diag_data += [X]

are_multisets_equal_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(0, 2, n)
    y = np.random.randint(0, 2, n)
    are_multisets_equal_data += [(x, y)]

max_prod_mod_3_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 100, n)
    max_prod_mod_3_data += [x]

convert_image_data = []
for n in [10, 100, 1000]:
    image = np.random.random((n, n, 10))
    weights = np.random.random(10)
    convert_image_data += [(image, weights)]

rle_scalar_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 3, (n, 2))
    y = np.random.randint(1, 3, (n, 2))
    shapes = np.random.randint(1, 3, n)
    x[:, 1] = shapes
    y[:, 1] = shapes

    rle_scalar_data += [(x, y)]

cosine_distance_data = []
for n in [10, 100]:
    X = np.random.random((n, n))
    Y = np.random.random((n, n))
    cosine_distance_data += [(X, Y)]
```

```python
# функция для подсчета времени работы

def count_time(func, *data):
    tick = time.perf_counter()
    func(*data)
    return time.perf_counter() - tick
```

```python
# получим время работы функций

# sum_non_neg_diag
sum_non_neg_diag_time = []
sum_non_neg_diag_vectorised_time = []

for X in sum_non_neg_diag_data:
  sum_non_neg_diag_time += [count_time(functions.sum_non_neg_diag, X.tolist())]
  sum_non_neg_diag_vectorised_time += [count_time(functions_vectorised.sum_non_neg_diag, X)]


# are_multisets_equal
are_multisets_equal_time = []
are_multisets_equal_vectorised_time = []

for x, y in are_multisets_equal_data:
  are_multisets_equal_time += [count_time(functions.are_multisets_equal, x.tolist(), y.tolist())]
  are_multisets_equal_vectorised_time += [count_time(functions_vectorised.are_multisets_equal, x, y)]


# max_prod_mod_3
max_prod_mod_3_time = []
max_prod_mod_3_vectorised_time = []

for x in max_prod_mod_3_data:
  max_prod_mod_3_time += [count_time(functions.max_prod_mod_3, x.tolist())]
  max_prod_mod_3_vectorised_time += [count_time(functions_vectorised.max_prod_mod_3, x)]


# convert_image
convert_image_time = []
convert_image_vectorised_time = []

for x, y in convert_image_data:
  convert_image_time += [count_time(functions.convert_image, x.tolist(), y.tolist())]
  convert_image_vectorised_time += [count_time(functions_vectorised.convert_image, x, y)]


# rle_scalar
rle_scalar_time = []
rle_scalar_vectorised_time = []

for x, y in rle_scalar_data:
  rle_scalar_time += [count_time(functions.rle_scalar, x.tolist(), y.tolist())]
  rle_scalar_vectorised_time += [count_time(functions_vectorised.rle_scalar, x, y)]


# cosine_distance
cosine_distance_time = []
cosine_distance_vectorised_time = []

for x, y in cosine_distance_data:
  cosine_distance_time += [count_time(functions.cosine_distance, x.tolist(), y.tolist())]
  cosine_distance_vectorised_time += [count_time(functions_vectorised.cosine_distance, x, y)]
```
