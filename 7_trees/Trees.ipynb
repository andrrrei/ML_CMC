`````python
import time
mean_absolute_error_train = np.zeros(10)
mean_absolute_error_test = np.zeros(10)
times = np.zeros(10)
for idx, item in enumerate(np.logspace(1, 4, num=10).astype(int)):
    forest = RandomForestRegressor(n_estimators=item, random_state=42)
    start = time.time()
    forest.fit(X_train, y_train)
    finish = time.time()
    times[idx] = finish - start
    mean_absolute_error_train[idx] = mean_absolute_error(y_train, forest.predict(X_train))
    mean_absolute_error_test[idx] = mean_absolute_error(y_test, forest.predict(X_test))
`````

`````python
f, (ax1, ax2) = plt.subplots(1, 2,figsize=(28, 14))
nums = np.logspace(1, 4, num=10).astype(int)
ax1.set_xlabel("n_estimators")
ax1.set_ylabel("times")
ax1.plot(nums, times)
ax1.grid(True)
ax2.plot(nums, mean_absolute_error_test, color='black', label = "test")
ax2.plot(nums, mean_absolute_error_train, color='orange', label = "train")
ax2.set_xlabel("n_estimators")
ax2.set_ylabel("MAE")
ax2.legend()
ax2.grid(True)
plt.show()
`````

`````python
MAE_test = []
times = []
forest = RandomForestRegressor(n_estimators=1000, warm_start=True, random_state=42)
start = time.time()
forest.fit(X_train, y_train)
finish = time.time()
times.append(finish - start)
MAE_test.append(mean_absolute_error(y_test, forest.predict(X_test)))
for n_estimators in range(2000, 10001, 1000):
    forest.n_estimators = n_estimators
    start = time.time()
    forest.fit(X_train, y_train)
    finish = time.time()
    times.append(finish - start)
    MAE_test.append(mean_absolute_error(y_test, forest.predict(X_test)))
`````

`````python
class PotentialTransformer:
    """
    A potential transformer.

    This class is used to convert the potential's 2d matrix to 1d vector of features.
    """

    def fit(self, x, y):
        """
        Build the transformer on the training set.
        :param x: list of potential's 2d matrices
        :param y: target values (can be ignored)
        :return: trained transformer
        """
        return self

    def fit_transform(self, x, y):
        """
        Build the transformer on the training set and return the transformed dataset (1d vectors).
        :param x: list of potential's 2d matrices
        :param y: target values (can be ignored)
        :return: transformed potentials (list of 1d vectors)
        """
        return self.transform(x)

    def transform(self, x):
        """
        Transform the list of potential's 2d matrices with the trained transformer.
        :param x: list of potential's 2d matrices
        :return: transformed potentials (list of 1d vectors)
        """
        return x.reshape((x.shape[0], -1))

# Your code here:
`````
